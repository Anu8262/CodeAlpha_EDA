```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sample_eda_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df['total_bill'], kde=True, color='skyblue')
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='sex', y='tip', data=df, hue='sex', palette='pastel', legend=False)
plt.title("Tip Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='day', y='tip', data=df, estimator=sum, hue='day', palette='Blues_r', legend=False)
plt.title("Total Tips by Day")
plt.xlabel("Day")
plt.ylabel("Total Tips")
plt.show()
```

```python
df['tip_percent'] = (df['tip'] / df['total_bill']) * 100
df[['total_bill', 'tip', 'tip_percent']].head()
```

```python
sns.histplot(df['tip_percent'], kde=True, color='lightgreen')
plt.title("Distribution of Tip Percentage")
plt.xlabel("Tip % of Total Bill")
plt.ylabel("Frequency")
plt.show()
```
